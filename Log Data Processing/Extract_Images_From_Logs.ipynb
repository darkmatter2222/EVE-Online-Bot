```python
from tqdm import tqdm
import json, os, errno
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://github.com/dhaitz/mplcyberpunk
import mplcyberpunk

plt.style.use("cyberpunk")
plt.rcParams['figure.facecolor'] = '#0d1117'
plt.rcParams['axes.facecolor'] = '#0d1117'
plt.rcParams['savefig.facecolor'] = '#0d1117'
```

```python
def mkdir_p(path):
    try:
        os.makedirs(path)
    except OSError as exc: # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else: raise
        
def copy(source, destination):
    with open(source, 'rb') as file:
        myFile = file.read()
    with open(destination, 'wb') as file:
        file.write(myFile)
```

```python
lines = ''
root_dir = '\\\\databrick\\N\\eve_logs'
#root_dir = "O:\\log_dir\\eve_miner"
with open(f"{root_dir}\\RedComputer_Main_UI.py_2023-04-13_19-44-33_579524.log") as f:
    lines = f.readlines()
```

```python
clsf_name = 'route_y_large_vert_class_v2'

clsf_lines = []
for line in tqdm(lines):
    if f"'clsf': '{clsf_name}'" in line:
        index = line.find('{')
        clsf_lines.append(json.loads(line[index:-1].replace("'", '"').replace('UUID("', '"').replace('")', '"').replace("True", '1').replace("False", '0')))
df = pd.DataFrame(clsf_lines)
df
```

```python
source = f'\\\\databrick\\N\\eve_logs\\{clsf_name}'
dest = r'O:\eve_models\training_data\unclass'
for i in df.index:
    this_source = f"{source}\\{df.loc[i, 'id']}.png"
    mkdir_p(f"{dest}\\{df.loc[i, 'class']}")
    this_dest = f"{dest}\\{df.loc[i, 'class']}\\{df.loc[i, 'id']}.png"
    copy(this_source, this_dest)
```
