```python
from tqdm import tqdm
import json
import pandas as pd
```

```python
lines = ''
with open(r"\\databrick\Active Server Drive\log_dir\eve_miner\miner_2023-03-11_11-24-37_649163.log") as f:
    lines = f.readlines()
```

```python
mining_class_lines = []
for line in tqdm(lines):
    if 'MiningBot.BotActions.Actions:mine_till_full_v2:311' in line:
        index = line.find('{')
        mining_class_lines.append(json.loads(line[index:-1].replace("'", '"').replace('UUID("', '"').replace('")', '"').replace("True", '1').replace("False", '0')))
```
